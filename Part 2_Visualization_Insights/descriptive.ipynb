```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
from datetime import datetime

# Create folder to save charts
os.makedirs('descriptive_charts', exist_ok=True)

print("=== DESCRIPTIVE ANALYSIS - FASHION E-COMMERCE FIRM ===\n")
```

```python
print("Loading data files...")

sales = pd.read_csv('sales.csv')                    # Revenue & COGS by day
orders = pd.read_csv('orders.csv')
order_items = pd.read_csv('order_items.csv')
products = pd.read_csv('products.csv')
customers = pd.read_csv('customers.csv')
returns = pd.read_csv('returns.csv')
web_traffic = pd.read_csv('web_traffic.csv')
inventory = pd.read_csv('inventory.csv')
geography = pd.read_csv('geography.csv')
```

```python
print(f"Total Products          : {products['product_id'].nunique():,}")
print(f"Product Categories      : {products['category'].nunique()}")
print(f"Segments                : {products['segment'].nunique()}")
print(f"Unique Sizes            : {products['size'].nunique()}")
print(f"Unique Colors           : {products['color'].nunique()}")

print(f"Total Customers         : {customers['customer_id'].nunique():,}")
print(f"Total Orders            : {orders['order_id'].nunique():,}")
print(f"Total Order Items       : {order_items.shape[0]:,}")

# Revenue summary
total_revenue = sales['Revenue'].sum()
total_days = sales['Date'].nunique()
avg_daily_revenue = sales['Revenue'].mean()
print(f"Total Revenue (2012-2022) : {total_revenue:,.2f} VND")
print(f"Average Daily Revenue     : {avg_daily_revenue:,.2f} VND")
print(f"Study Period              : {sales['Date'].min()} → {sales['Date'].max()}\n")
```

```python
print("\n👥 3. CUSTOMER PROFILE")

age_summary = customers['age_group'].value_counts(normalize=True).round(3)*100
print("Age Group Distribution (%):\n", age_summary)

gender_summary = customers['gender'].value_counts(normalize=True).round(3)*100
print("\nGender Distribution (%):\n", gender_summary)

acq_summary = customers['acquisition_channel'].value_counts().head(8)
print("\nTop Acquisition Channels:\n", acq_summary)
```

```python
print("\n📦 4. ORDER & RETURN OVERVIEW")

return_rate = (returns.shape[0] / order_items.shape[0]) * 100
print(f"Overall Return Rate     : {return_rate:.2f}%")

return_by_reason = returns['return_reason'].value_counts(normalize=True).round(3)*100
print("\nTop Return Reasons (%):\n", return_by_reason)
```

```python
print("\n🌐 5. WEBSITE TRAFFIC OVERVIEW")

traffic_summary = web_traffic.groupby('traffic_source').agg(
    sessions=('sessions', 'sum'),
    unique_visitors=('unique_visitors', 'sum'),
    bounce_rate=('bounce_rate', 'mean')
).round(3)
traffic_summary['bounce_rate'] = traffic_summary['bounce_rate'] * 100
print(traffic_summary)
# Save chart for your friend
#fig2 = px.bar(traffic_summary.reset_index(), x='traffic_source', y='sessions',
              #title='Total Sessions by Traffic Source',
              #labels={'sessions': 'Total Sessions'})
#fig2.write_image('descriptive_charts/02_traffic_by_source.png')
#fig2.show()
```

```python
print("\n📦 6. INVENTORY HEALTH (Latest Snapshot)")

latest_inventory = inventory[inventory['snapshot_date'] == inventory['snapshot_date'].max()]
print(f"Latest snapshot date    : {latest_inventory['snapshot_date'].iloc[0]}")
print(f"Total products in stock : {latest_inventory['stock_on_hand'].sum():,}")
print(f"Products with stockout  : {(latest_inventory['stockout_flag'] == 1).sum()}")
```

```python
print("📦 2. PRODUCT PORTFOLIO")

cat_summary = products.groupby('category').agg(
    products=('product_id','nunique'),
    avg_price=('price','mean'),
    avg_cogs=('cogs','mean')
).round(2)
print(cat_summary)

# Save chart
#fig = px.bar(cat_summary.reset_index(), x='category', y='products', 
             #title='Number of Products by Category',
             #labels={'products': 'Number of Products'})
#fig.write_image('descriptive_charts/01_products_by_category.png')
```

```python
summary = {
    "Total_Revenue": total_revenue,
    "Avg_Daily_Revenue": avg_daily_revenue,
    "Total_Orders": orders['order_id'].nunique(),
    "Total_Customers": customers['customer_id'].nunique(),
    "Overall_Return_Rate_%": return_rate,
    "Study_Period": f"{sales['Date'].min()} to {sales['Date'].max()}"
}

pd.Series(summary).to_csv('descriptive_charts/company_summary.csv')
print("\n✅ All descriptive analysis completed! Charts saved in 'descriptive_charts' folder.")
print("✅ Summary saved as 'company_summary.csv'")
```
